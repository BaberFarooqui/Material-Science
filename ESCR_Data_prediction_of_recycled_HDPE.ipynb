```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Define the bottle types and PCR content range
bottle_types = ['Type A', 'Type B', 'Type C', 'Type D']
pcr_content = np.linspace(0, 40, 5)  # PCR content from 0% to 40%

# Generate synthetic molecular weights and time to failure based on PCR content
data = []
for bottle in bottle_types:
    for pcr in pcr_content:
        # Simulate a decrease in molecular weight with an increase in PCR content
        mw = 200 - 1.5 * pcr + np.random.normal(0, 5)  # Linear decrease with some noise
        # Simulate time to failure with some randomness
        time_to_failure = mw * (1 - pcr / 100) + np.random.normal(0, 10)
        data.append([bottle, pcr, mw, time_to_failure])

# Create a DataFrame
df = pd.DataFrame(data, columns=['Bottle Type', 'PCR Content (%)', 'Molecular Weight (kg/mol)', 'Time to Failure (hours)'])

# Plotting Time to Failure vs. Molecular Weight
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Molecular Weight (kg/mol)', y='Time to Failure (hours)', hue='Bottle Type', style='Bottle Type', s=100)

# Customize the plot
plt.title('Time to Failure vs. Molecular Weight by Bottle Type')
plt.xlabel('Molecular Weight (kg/mol)')
plt.ylabel('Time to Failure (hours)')
plt.legend(title='Bottle Type')
plt.grid(True)
plt.show()

# Plotting PCR Content vs. Molecular Weight
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='PCR Content (%)', y='Molecular Weight (kg/mol)', hue='Bottle Type', marker='o')

# Customize the plot
plt.title('PCR Content vs. Molecular Weight by Bottle Type')
plt.xlabel('PCR Content (%)')
plt.ylabel('Molecular Weight (kg/mol)')
plt.legend(title='Bottle Type')
plt.grid(True)
plt.show()

# Display the DataFrame
print(df)

```
